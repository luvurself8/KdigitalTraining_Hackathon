```python
def boat (num, busan_drag_train):
    boat = busan_drag_train[busan_drag_train['선박번호']==num]
    for i in ['위도', '경도', '대지속력', '실침로', '선수미선','점사이_거리', '평균선회반경(m)', '유속', '유향', '유행', '풍향', '풍속','최대파고', '유의파고', '평균파고']:
        boat[i].plot()
        boat[boat['answer']==num][i].plot(color='red')
        plt.title(i)
        plt.show()

    
```
